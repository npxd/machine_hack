```
from tqdm import tqdm_notebook as tqdm
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
%matplotlib inline
from keras.preprocessing.text import text_to_word_sequence, Tokenizer
from keras.preprocessing.sequence import pad_sequences
custom_style = {'axes.labelcolor': 'white',
                'xtick.color': 'white',
                'ytick.color': 'white'}
sns.set_style(rc=custom_style)
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stopwords = set(stopwords.words('english'))
t=Tokenizer()
```

### Add custom stopwords here

```
for i in ',', '"', "'", ":", ";", ":", "’":  stopwords.add(i)
```

### Importing df

```
df = pd.read_excel("Data_Train.xlsx")
df
```

```
target = df['SECTION'].values
df.drop(columns=['SECTION'], inplace=True)
```

### Data Prep

```
# tokenize and convert to lowercase. keras does it all :)
# also remove stopwords, this isnt there in keras i guess
# using sets to remove redundant words
# df['tokenized'] = [[word_tokenize(sentence) for sentence in para.split('.')] for para in tqdm(df.STORY)]
df['tokenized'] = [(list(word for word in text_to_word_sequence(para) if word not in stopwords)) for para in tqdm(df.STORY)]
df['tokenized'][0]
```

```
# Revert to stemmer since thats more importing (I think)
# removing the words for which the stems are the same since that eliminates the proper nouns - this is needed for embeddings with GloVe
ps = PorterStemmer()
# wnl = WordNetLemmatizer()
df['stemmed'] = [" ".join(list(ps.stem(word) for word in words if ps.stem(word) != word)) for words in tqdm(df['tokenized'])]
```

```
df['stemmed'][0]
```

### Para 1 experiment

```
t.fit_on_texts(df.STORY.values)
max_features = 0
max_length = 0
def encode(df):
  global max_features, max_length
  encoded_docs = t.texts_to_sequences(df.STORY.values)
  matrix = t.texts_to_matrix(df.STORY.values)
  max_features = matrix.shape[1]
  max_length = 500
  return pad_sequences(encoded_docs, maxlen=max_length, padding='post')
padded_docs = encode(df)
```

```
print(padded_docs.shape)
print(np.unique(target))
```

### Model

```
from keras.models import Sequential
from keras.layers import Embedding, Dense, LSTM
from keras.utils import to_categorical
from keras.callbacks import EarlyStopping

early_stopping_monitor = EarlyStopping(patience=1)
y = to_categorical(target, num_classes=4)
print(y)
model = Sequential()
model.add(Embedding(max_features, 32, input_length=max_length))
model.add(LSTM(32))
model.add(Dense(4, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])

h = model.fit(padded_docs, y, epochs=10, validation_split=0.2, callbacks=[early_stopping_monitor])

```

```
test_df = pd.read_excel('test.xlsx')
padded = encode(test_df)

```

```
pred = model.predict(padded)
```
